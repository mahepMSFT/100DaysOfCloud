```python
import pandas as pd

df = pd.read_csv("student.csv")

df["Total"] = df["S1"] + df["S2"] + df["S3"]

print(df.head())
```

```python
print(df.columns)
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df["Percentage"] = (df["Total"]/300)*100

print(df.head())
```

```python
def grade(p):
    if p >= 80:
        return "A"
    elif p >= 60:
        return "B"
    elif p >= 40:
        return "C"
    else:
        return "FAIL"

df["Grade"] = df["Percentage"].apply(grade)

print(df.head())
```

```python
import matplotlib.pyplot as plt

gender_count = df["Gender"].value_counts()
gender_percentage = (gender_count / len(df)) * 100
plt.bar(gender_percentage.index, gender_percentage.values)

plt.title("Percentage of Male vs Female Students")
plt.xlabel("Gender")
plt.ylabel("Percentage")

plt.show()
```

```python
df["Result"] = df["Grade"].apply(lambda x: "PASS" if x != "FAIL" else "FAIL")
print(df[["Name","Grade","Result"]])
```

```python
result_count = df["Result"].value_counts()
plt.pie(result_count.values,
        labels=result_count.index,
        autopct="%1.1f%%")

plt.title("PASS vs FAIL Percentage")

plt.show()
```
